## labeling_instructions

This notebook presents examples of correct landmark labeling in a rat. Use the following table of contents to navigate quickly to your marker of choice. **Make sure to look at the "quickstart" reference images below the table of contents as well.**

Note that while these example animals have "markers" (beads) attached at each landmark position, the images you are labeling will not have these markers -- have fun!

[HeadF](#HeadF). The right front corner of the head mount/cap.

[HeadBR](#HeadBR). The back/bottom right corner of the head mount/cap.

[HeadBL](#HeadBL). The back/bottom left corner of the head mount/cap.

[SpineF](#SpineF). The front of the spine (anterior thoracic), between the left and ride shoulders.

[SpineM](#SpineM). The middle of the spine (anterior lumbar / posterior thoracic).

[SpineB](#SpineB). The back/bottom of the spine (sacral), beetween the two hips.

[Offset1](#Offset1). To the left of the spine, between the anterior/posterior positions of SpineF and SpineM.

[Offset2](#Offset2). To the left of the spine, between the anterior/posterior positions of SpineM and SpineL.

[HipL](#HipL). Top of the left leg. Posterior to SpineL.

[HipR](#HipR). Top of the right leg. Posterior to SpineL.

[ElbowL](#ElbowL). Left elbow. At the middle of the left arm (forelimb).

[ArmL](#ArmL). Left forearm (forelimb).

[ShoulderL](#ShoulderL). Left Shoulder. To the left of SpineF.

[ShoulderR](#ShoulderR). Right Shoulder. To the right of SpineF.

[ElbowR](#ElbowR). Right elbow. At the middle of the right arm (forelimb).

[ArmR](#ArmR). Right forearm (forelimb).

[KneeR](#KneeR). Above the right knee. Anterior to HipR.

[KneeL](#KneeL). Above the left knee. Anterior to HipL.

[ShinL](#ShinL). On the lower end of the left leg (hindlimb).

[ShinR](#ShinR). On the lower end of the right leg (hindlimb).
